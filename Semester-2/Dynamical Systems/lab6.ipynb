```python
from sage.plot.line import Line
N = 100
lambd = 4
x = zero_vector(RR,N+1)
y = zero_vector(RR,N+1)
x[0]=0.67
y[0]=0.67001
for n in range(1,N):
    x[n] = lambd * x[n-1] * (1 - x[n-1])
    y[n] = lambd * y[n-1] * (1 - y[n-1])
lx = line(Line(range(N),x,{}), color='blue',marker='o')
ly = line(Line(range(N),y,{}), color='red',marker='o')
show(lx+ly)
```

```python
N=11
h=1/(N-1)
x = zero_vector(RR,N)
y = zero_vector(RR,N)
y[0]=1
x[0] = 0
ye = zero_vector(RR,N)
ye[0] = 1
for i in range(1,N):
    x[i] = x[i-1] + h
    y[i] = y[i-1] + h* 2 * x[i-1] * y[i-1]
    ye[i] = exp(x[i]^2)
le = line(Line(x,y,{}), color='blue',marker='o')
lye = line(Line(x,ye,{}), color='red',marker='o')
show(le+lye)
```

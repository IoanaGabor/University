```python
var('t')
x = function('x')(t)
eq = (diff(x,t,2) + 3*diff(x,t,1) + x == 1)
sol = desolve(eq,x)
limit(sol, t=infinity)
```

```python
var('t')
x = function('x')(t)
eq = (diff(x,t,2) + 4*x == 1)
sol(t) = desolve(eq,x,ics=[0,5/4,0])
sol(pi)
```

```python
var('t')
x = function('x')(t)
f1 = function('f1')(t)
f2 = function('f2')(t)
f1(t) = t
f2(t) = pi*exp(pi-t)
eq1 = (diff(x,t,2) + x == f1)
eq2 = (diff(x,t,2) + x == f2)
sol1(t) = desolve(eq1,x,ics=[0,0,1])
sol2(t) = desolve(eq2,x,ics=[pi,sol1.subs(t=pi),diff(sol1,t,1).subs(t=pi)])
sol = piecewise([((0,pi),sol1),((pi,infinity),sol2)])
plot(sol,[0,10])
```
